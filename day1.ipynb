```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

from sklearn import svm
sm = svm.SVC()

from sklearn import tree
tr = tree.DecisionTreeClassifier()
```
