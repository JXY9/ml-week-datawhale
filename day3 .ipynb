## 基本库

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/dell/Desktop/day1_data_all.csv')
X = data.drop(['status'],axis =1)
Y = data.loc[:,'status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3,random_state = 2018)
```

## 评分相关

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
```

# 逻辑回归 

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
lr_acc = accuracy_score(y_test, y_pred)
auc_lr = roc_auc_score(y_test, y_pred)
print(classification_report(y_test, y_pred))  # target_names=target_names
```

# SVM

```python
from sklearn import svm
sm = svm.SVC()
sm.fit(x_train,y_train)
y_pred = sm.predict(x_test)
sm_acc = accuracy_score(y_test, y_pred)
auc_sm = roc_auc_score(y_test, y_pred)
print(classification_report(y_test, y_pred))
```

# 决策树

```python
from sklearn import tree
tr = tree.DecisionTreeClassifier()
tr.fit(x_train,y_train)
y_pred = tr.predict(x_test)
tr_acc = accuracy_score(y_test, y_pred)
auc_tr = roc_auc_score(y_test, y_pred)
print(classification_report(y_test, y_pred))
```

# 随机森林

```python
from sklearn.ensemble import RandomForestClassifier 
rfc = RandomForestClassifier()
rfc = rfc.fit(x_train,y_train)
y_pred = rfc.predict(x_test)
rfc_acc = accuracy_score(y_test, y_pred)
auc_rfc = roc_auc_score(y_test, y_pred)
print(classification_report(y_test, y_pred))
```

# GBDT

```python
from sklearn.ensemble import GradientBoostingClassifier 
gbdtc = GradientBoostingClassifier()
gbdtc.fit(x_train,y_train)
y_pred = gbdtc.predict(x_test)
gbdtc_acc = accuracy_score(y_test, y_pred)
auc_gbdtc = roc_auc_score(y_test, y_pred)
print(classification_report(y_test, y_pred))
```

# XGBoost

```python
from xgboost.sklearn import XGBClassifier 

xgbc =  XGBClassifier(random_state = 2018)
xgbc.fit(x_train,y_train)  # xgbc.score(x_test,y_test)
y_pred = xgbc.predict(x_test)
xgbc_acc = accuracy_score(y_test, y_pred)
auc_xgbc = roc_auc_score(y_test, y_pred)
print(classification_report(y_test, y_pred))
```

# LightGBM

```python
import lightgbm as lgb
lgbc = lgb.LGBMClassifier()
lgbc.fit(x_train,y_train)
y_pred = lgbc.predict(x_test)
lgbc_acc = accuracy_score(y_test, y_pred)
auc_lgbc = roc_auc_score(y_test, y_pred)
print(classification_report(y_test, y_pred))
```
